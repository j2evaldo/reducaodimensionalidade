## **Sobre as cores das Imagens**

É importante compreender que as imagens RGB são compostas por três dimensões é que através da escala de cada uma destas dimensões é que a coloração da imagem é definida.

Cada dimensão destas, vai de 0 a 255. E a medida que esta variação vai sendo alterada, a cor da imagem também o é.  

Quando estamos trabalhando com programas de manipulação de cores de imagens, o que estamos fazendo é basicamente alterar estas escalas de forma individual. 


---

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
# Importando a Imagem
imagem = 'colorida.jpg'

# Carregando a Imagem para que possa exibi-la com o Matplotlib
temp = plt.imread(imagem)   

# Verificando as dimensões da Imagem
temp.shape
```

```python
plt.imshow(temp);
```

## **Sobre a escala de cinza**

Quando quero transformar uma imagem em escala de cinza eu basicamente tenho que fazer a média da escala de todas as camadas e aplicar o resultado deste cálculo em cada uma. 
Desta forma não terei variação, uma vez que todas terão o mesmo valor de escala.


```python
imagem1 = Image.open('colorida.jpg')
matriz = imagem1.load()

# Conversão da Imagem em escala de Cinza
for i in range(imagem1.size[0]):
    for x in range(imagem1.size[1]):
        # Armazenando o valor de escala de cada camada
        r = matriz[i, x][0]
        g = matriz[i, x][1]
        b = matriz[i, x][2]
    # Calculando o valor médio da escala
        media = int((r + g + b) / 3)
        matriz[i,x] = (media, media, media)

# Salvando a Imagem em Escala de Cinza
imagem1.save('escalacinza.jpg')

# Exibindo a Imagem em Escala de Cinza
imagem = 'escalacinza.jpg'
temp = plt.imread(imagem)  
plt.imshow(temp);
```

## **Convertendo Imagem em Preto e Branco**

Agora que já tenho a imagem em escala de cinza, minha vida fica um pouco mais fácil porque basta eu trabalhar em cima de duas camadas jogando os valores para 0 ou 1. 

Mas antes disso tenho que aplicar o método .dot do numpy que me retornará o produto escalar dos vetores de entrada. Se ambas as entradas forem escalares, ele produzirá um array 1-D, caso contrário, um array n-dimensional.

Neste caso, acabo recebendo um array n-dimensional e comparo todos os valores com 120. Caso algum dos valores seja maior do que 120, recebo um True com retorno, caso não, recebo um False. 

Como estamos falando de Python, posso tranquilamente pegar este retorno e realizar algumas operações matemáticas. É o caso por exemplo do isnull().sum(), na prática estou somando uma matriz de verdadeiros ou falsos. 

Para este problema, o que faço é uma multiplicação onde terei como resultado um array n-dimensional de zeros e uns que será passado para o método im.show do Matplotlib. 


```python
#Convertendo imagem Preto e Braco
import matplotlib.image as mpimg
img = mpimg.imread("escalacinza.jpg")
img = np.dot(img[...,:2], [0.3, 0.6])
temp = 1 * (img > 120)

plt.imshow(temp, cmap= plt.get_cmap(name = 'gray'));
```
